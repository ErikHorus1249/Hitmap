```python
import pandas as pd
import seaborn
```

```python
df = pd.read_excel("sample.xlsx", index_col=0)
```

```python
df
```

```python
seaborn.heatmap(df, cmap="coolwarm")
```

```python
seaborn.heatmap(df, cmap="coolwarm", xticklabels=[])
```

```python
seaborn.heatmap(df, cmap="coolwarm", xticklabels=[]).set_title("Average manual temperature in the UK, 1984-2021")
```

```python
seaborn.heatmap(df, cmap="coolwarm", xticklabels=[], cbar_kws={"label": "Temperature [Độ C]"}).set_title("Average manual temperature in the UK, 1984-2021")
```

```python
# https://python-graph-gallery.com/92-control-color-in-seaborn-heatmaps/
```

```python
seaborn.heatmap(df, cmap="Greens", xticklabels=[], cbar_kws={"label": "Temperature [Độ C]"}).set_title("Average manual temperature in the UK, 1984-2021")
```
