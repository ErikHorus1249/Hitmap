# Introduction to Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

## Numberical Plots
+ Distribution: histogram, KDE(Kernel Density Estimate)
+ Bar plot
+ Count plot
+ Facet Plot
+ Box Plot
+ Join Plot
+ Pair Plot
+ Heatmap

```python
## Styling
```

```python
sns.set_theme()
```

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head()
```

```python
sns.histplot(data=tips_df["total_bill"])
```

```python
tips_df["total_bill"].value_counts().sort_values(ascending=False)
```

### KDE Plot

```python
sns.kdeplot(data=tips_df["total_bill"])
```

### Displot

```python
sns.displot(data=tips_df, x="total_bill", col="time", kde=True)
```

## Bar plot

```python
sns.barplot(data=tips_df, x="sex", y="tip", estimator=np.mean)
```

### Count plot

```python
tips_df["sex"].value_counts()
```

```python
sns.countplot(data=tips_df, x="sex")
```

### Box plot

```python
sns.boxplot(data=tips_df, x="day", y="total_bill", hue="sex", palette="Blues")
plt.legend(loc=0)
```

### Facet Grid

```python
tips_df.head()
```

```python
tips_fg = sns.FacetGrid(data=tips_df, row="smoker", col="time") # create facet class instance 
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
sns.scatterplot(data=tips_df,x='total_bill', y='tip')
```

```python
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       row="time",
                       hue="smoker",
                      col_order=["Female", "Male"],
                      palette="Set2",
                      height=4,
                      aspect=1.5)
new_fg.map(sns.scatterplot, "total_bill", "tip", s=100, edgecolor='b', alpha=1)
new_fg.add_legend()
```

### Joint Plot

```python
penguins_df = sns.load_dataset("penguins")
```

```python
penguins_df.head()
```

```python
penguins_df["island"].value_counts()
```

```python
sns.jointplot(data=penguins_df, x="bill_length_mm", y="flipper_length_mm", hue="species")
```

### Pair Plot

```python
sns.pairplot(data=penguins_df, hue="species") 
```

### Heatmap

```python
flights_df = sns.load_dataset("flights")
```

```python
flights_df.head()
```

```python
flights = pd.pivot_table(flights_df, index='month', columns='year', values='passengers')
```

```python
flights
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
```
