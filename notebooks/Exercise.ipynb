```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 01.Load dataset

```python
cereal_df = pd.read_csv("../data/cereal.csv")
```

```python
cereal_df.head()
```

# 02.Irrelevant Fields

```python
fields = ["shelf", "weight", "cups", "rating"]
new_cereal_df = cereal_df.drop(fields, axis=1)
new_cereal_df.head()
```

# 03.Correlation

```python
cereal_corr = new_cereal_df.corr(numeric_only=True) ## get correlation from data frame
cereal_corr
```

## One_likes can build a matrix of boolean (True, False) with the same shape as your data

```python
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

## Np's triu: return only upper triangle matrix

```python
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask)
```

```python
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
adjusted_cereal_corr
```

```python
fig, ax = plt.subplots(figsize=(10,8))

# hsluv.org
# cmap = sns.diverging_palette(347.7, 76.7, 58.4, as_cmap=True)
cmap = sns.color_palette("viridis", as_cmap=True)

sns.heatmap(data=adjusted_cereal_corr, 
            mask=adjusted_mask, 
            annot=True, 
            annot_kws={"fontsize": 12},
            fmt=".2f", 
            cmap=cmap,
            vmin=-1, 
            vmax=1, linecolor='white', linewidths=.5)

yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=12)
ax.set_xticklabels(xticks, rotation=90, fontsize=12)

title = "CORRELATION MATRIX \nSAMPLED CEREALS COMPOSITIONS\n"
ax.set_title(title, loc='left', fontsize=18)
```
